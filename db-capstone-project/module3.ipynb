```python
import mysql.connector as connector

connection = connector.connect(
    user="vivek",
    password="root",
    db="littlelemondb"
)

cursor = connection.cursor()

print("Connected to database successfully!")
```

```python
show_tables_query = "SHOW TABLES"
cursor.execute(show_tables_query)

results = cursor.fetchall()

print(results)
```

```python
query = """
SELECT 
    Customers.CustomerName, 
    Customers.ContactNumber, 
    Customers.Email, 
    SUM(Orders.TotalCost) AS BillAmount
FROM Customers
INNER JOIN Bookings ON Customers.CustomerID = Bookings.CustomerID
INNER JOIN Orders ON Bookings.BookingID = Orders.BookingID
GROUP BY Customers.CustomerName, Customers.ContactNumber, Customers.Email
HAVING BillAmount > 60;
"""

cursor.execute(query)
results = cursor.fetchall()

print("Customers who spent more than $60:")
for row in results:
    print(row)
```

```python
cursor.close()
connection.close()
print("Connection closed.")
```
